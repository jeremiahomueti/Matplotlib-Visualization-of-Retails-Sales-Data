## Question 1: Bar Chart

**Task:** Create a bar chart showing the total sales for each region.

**Requirements:**
- Use a bar chart
- Add a title and axis labels

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = "retail.xlsx"   # make sure the file is in the same folder as your notebook
df = pd.read_excel(file_path)

# Drop unnecessary index column if it exists
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

# Group data by region and sum the sales
sales_by_region = df.groupby('region')['sales'].sum()

# Create bar chart
plt.figure(figsize=(8, 5))
sales_by_region.plot(kind='bar', color='skyblue', edgecolor='black')

# Add title and axis labels
plt.title('Total Sales by Region', fontsize=14, weight='bold')
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

# Add grid and improve layout
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```


## Question 2: Horizontal Bar Chart

**Task:** Create a horizontal bar chart showing the average profit for each product category.

**Requirements:**
- Use `plt.barh()`
- Add a title and axis labels

```python
# Group data by product category and calculate average profit
avg_profit_by_category = df.groupby('product_category')['profit'].mean().sort_values()

# Create horizontal bar chart
plt.figure(figsize=(8, 5))
plt.barh(avg_profit_by_category.index, avg_profit_by_category.values, color='lightgreen', edgecolor='black')

# Add title and axis labels
plt.title('Average Profit by Product Category', fontsize=14, weight='bold')
plt.xlabel('Average Profit', fontsize=12)
plt.ylabel('Product Category', fontsize=12)

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```


## Question 3: Pie Chart

**Task:** Create a pie chart showing the distribution of transactions by store type.

**Requirements:**
- Show percentages on the chart
- Add a title

```python
# Your code here
# Count the number of transactions by store type
transactions_by_store = df['store_type'].value_counts()

# Create pie chart
plt.figure(figsize=(7, 7))
plt.pie(
    transactions_by_store,
    labels=transactions_by_store.index,
    autopct='%1.1f%%',     # show percentages
    startangle=90,         # rotate start for better alignment
    colors=plt.cm.Paired.colors,  # nice preset color palette
    wedgeprops={'edgecolor': 'black'}  # outline slices
)

# Add title
plt.title('Distribution of Transactions by Store Type', fontsize=14, weight='bold')

# Make it look clean and round
plt.axis('equal')  # ensures pie is circular
plt.tight_layout()

# Display chart
plt.show()
```


## Question 4: Scatter Plot

**Task:** Create a scatter plot showing the relationship between sales and profit.

**Requirements:**
- Sales on x-axis, profit on y-axis
- Add a title and axis labels

```python
# Create scatter plot: Sales (x-axis) vs Profit (y-axis)
plt.figure(figsize=(8, 6))
plt.scatter(df['sales'], df['profit'], color='teal', alpha=0.6, edgecolor='black')

# Add title and axis labels
plt.title('Relationship Between Sales and Profit', fontsize=14, weight='bold')
plt.xlabel('Sales', fontsize=12)
plt.ylabel('Profit', fontsize=12)

# Add grid for readability
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Show chart
plt.show()
```


## Question 5: Line Chart

**Task:** Create a line chart showing total sales for each month (1-12).

**Requirements:**
- Use `plt.plot()` with markers
- Add a title and axis labels

```python
# Group data by month and sum the sales
monthly_sales = df.groupby('month')['sales'].sum().sort_index()

# Create line chart
plt.figure(figsize=(8, 5))
plt.plot(
    monthly_sales.index, 
    monthly_sales.values, 
    marker='o',             # adds circle markers
    linestyle='-', 
    color='royalblue', 
    linewidth=2
)

# Add title and axis labels
plt.title('Total Sales by Month', fontsize=14, weight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

# Add grid for clarity
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Show chart
plt.show()
```


## Question 6: Histogram


## Question 7: Multiple Lines

**Task:** Create a line chart showing both sales and profit by month on the same plot.

**Requirements:**
- Two lines (sales and profit)
- Add a legend
- Add a title and axis labels

```python
# Group data by month and sum both sales and profit
monthly_data = df.groupby('month')[['sales', 'profit']].sum()

# Create the line chart
plt.figure(figsize=(10, 6))

# Plot sales
plt.plot(monthly_data.index, monthly_data['sales'], linewidth=2, label='Sales')

# Plot profit
plt.plot(monthly_data.index, monthly_data['profit'], linewidth=2, label='Profit')

# Add title and axis labels
plt.title('Sales and Profit by Month', fontsize=14, weight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Value', fontsize=12)

# Add legend
plt.legend()

# Improve layout
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show plot
plt.show()
```


## Question 9: Subplots

**Task:** Create a figure with 2 subplots side by side:
- Left: Bar chart of sales by product category
- Right: Pie chart of customer gender distribution

**Requirements:**
- Use `plt.subplots(1, 2)`
- Add titles to both subplots

```python
# --- Prepare the data ---

# Sales by product category
sales_by_category = df.groupby('product_category')['sales'].sum()

# Customer gender distribution
gender_counts = df['customer_gender'].value_counts()

# --- Create subplots ---
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# ----- LEFT SUBPLOT: Bar Chart -----
axes[0].bar(sales_by_category.index, sales_by_category.values)
axes[0].set_title('Sales by Product Category', fontsize=13, weight='bold')
axes[0].set_xlabel('Product Category')
axes[0].set_ylabel('Total Sales')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# ----- RIGHT SUBPLOT: Pie Chart -----
axes[1].pie(gender_counts.values, labels=gender_counts.index, autopct='%1.1f%%')
axes[1].set_title('Customer Gender Distribution', fontsize=13, weight='bold')

# Improve layout
plt.tight_layout()

# Show the figure
plt.show()
```


## Question 10: Scatter Plot with Colors

**Task:** Create a scatter plot of quantity vs sales, where points are colored by region.

**Requirements:**
- Use different colors for each region
- Add a legend
- Add a title and axis labels

**Hint:** You can plot each region separately in a loop.

```python
# Group data by region
regions = df['region'].unique()

# Create scatter plot
plt.figure(figsize=(10, 6))

for region in regions:
    subset = df[df['region'] == region]
    plt.scatter(subset['quantity'], subset['sales'], label=region)

# Add title and axis labels
plt.title('Quantity vs Sales Colored by Region', fontsize=14, weight='bold')
plt.xlabel('Quantity', fontsize=12)
plt.ylabel('Sales', fontsize=12)

# Add legend
plt.legend(title='Region')

# Improve layout
plt.grid(linestyle='--', alpha=0.7)
plt.tight_layout()

# Show plot
plt.show()
```
